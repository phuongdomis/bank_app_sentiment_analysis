```python
# import emoji dictionary
emojicon_dict = {}
emojicon_file = open("emojicon.txt")
for line in emojicon_file:
    key, value = line.split('\t')
    emojicon_dict[key] = value.replace('\n', '')

# import file teencode
teencode_dict = {}
teencode_file = open("teencode.txt")
for line in teencode_file:
    key, value = line.split('\t')
    teencode_dict[key] = value.replace('\n', '')
```

```python
from underthesea import word_tokenize
```

```python
import pickle
def Load_Object(filename):
    with open(filename, 'rb') as file:
        obj = pickle.load(file)
    return obj
```

```python
my_model_vnm = Load_Object('bank_app_LogisticRegression_vnm.pkl')
count_vnm = Load_Object('Count_tfidf_vnm.pkl')
```

```python
def xu_ly(text):
  text = convert_emoticons(text, EMOTICONS_EMO)
  text = convert_emoji(text, emojicon_dict)
  text = convert_teencode(text, teencode_dict)
  text = remove_special_icons(text)
  text = convert_khong_words(text, final_ko_list)
  text = clean_text(text)
  text = word_tokenize(text, format='text')
  return text
```

```python
def predict(x_new):
  x_new = xu_ly(x_new)
  x_new = count_vnm.transform(np.array([x_new])).toarray()
  y_prob = my_model_vnm.predict_proba(x_new)
  if y_prob[0][0] >= 0.5:
    reply = 'Không thích/ Dislike'
  else:
    reply = 'Thích/ Like'
  return reply
```

```python
my_model_eng = Load_Object('bank_app_LogisticRegression_eng.pkl')
count_eng = Load_Object('Count_tfidf_eng.pkl')
```

```python
def process(text):
  text = emoji.demojize(text)
  text = remove_special_icons(text)
  text = clean_text(text)
  return text
```

```python
def predict_eng(x_new):
  x_new = process(x_new)
  x_new = count_eng.transform(np.array([x_new])).toarray()
  y_prob = my_model_eng.predict_proba(x_new)
  if y_prob[0][0] >= 0.5:
    reply = 'Không thích/ Dislike'
  else:
    reply = 'Thích/ Like'
  return reply
```

```python
from langdetect import detect
```

```python
def check_is_en(text):
    try:
        language = detect(text)
    except:
        language = "error"
    return (language == 'en')
```

```python
def predict_reviews(text):
  if check_is_en(text):
    result = predict_eng(text)
  else:
    result = predict(text)
  return result
```
